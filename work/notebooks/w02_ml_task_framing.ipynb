## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*



My chosen lane focuses on identifying webpages that should be prioritized for content refresh.

This is primarily a ranking and classification problem. The system ranks pages based on their likelihood of declining performance so editors can decide which pages should be updated first.

## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*



The target is whether a webpage is declining.

The label is created from the observed value in the `trend_direction` column.

A page with `trend_direction = "down"` is labeled as declining.

This is an observed outcome available in the starter dataset.

## 3. Success metric

*One metric you can defend. What number means 'good'?*



The success metric is Precision@50.

A higher Precision@50 means that more of the pages recommended by the model are actually declining, allowing editors to focus on the most important pages first.

## 4. The unit of analysis, as a real dataframe

*Load your lane's slice and show it: one row = one what?*

```python
import pandas as pd

df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

df["is_declining_label"] = (
    df["trend_direction"].str.lower().eq("down").astype(int)
)

print("Rows:", len(df))
print("Columns:", len(df.columns))

df[[
    "content_id",
    "content_type",
    "impressions_90d",
    "ctr",
    "word_count",
    "trend_direction",
    "is_declining_label"
]].head()
```

## Unit of Analysis

Each row represents one webpage.

The model analyzes search performance, engagement, and content characteristics for each page to determine whether it should be prioritized for review.

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*



A simple rule can only consider a few conditions at a time, such as impressions or content age.

Machine learning can evaluate multiple signals simultaneously, including CTR, search performance, ranking position, content age, and engagement metrics.

This helps prioritize pages more accurately while supporting editorial decisions.

The results should be interpreted as observed and decision-support, not as causal proof.
